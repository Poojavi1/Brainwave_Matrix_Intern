```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the sales data from the CSV file
df = pd.read_csv("C:/Users/91901/Downloads/Stores.csv")
```

```python
# Show the first few rows of the data
print("Sales Data:")
print(df.head())
```

```python
# Basic Statistics
print("\nBasic Statistics:")
print(df.describe())
```

```python
# Total sales for each product
print("\nStore Sales per Product:")
total_sales_per_product = df.groupby('Items_Available')['Store_Sales'].sum()
print(total_sales_per_product)
```

```python
# Total sales for each day
print("\nStore Sales per Day:")
df['Date'] = pd.to_datetime(df['Store_Sales'])  # Convert Date to datetime
total_sales_per_day = df.groupby('Date')['Store_Sales'].sum()
print(total_sales_per_day)
```

```python
# Plotting Total Sales per Product
plt.figure(figsize=(10, 6))
total_sales_per_product.plot(kind='bar', color='skyblue')
plt.title('Store Sales per Product')
plt.xlabel('Items_Available')
plt.ylabel('Store_Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Plotting Total Sales per Day
plt.figure(figsize=(10, 6))
total_sales_per_day.plot(kind='line', marker='o', color='orange')
plt.title('Store Sales per Day')
plt.xlabel('Date')
plt.ylabel('Store_Sales')
plt.tight_layout()
plt.show()
```
